### Using unsupervised machine learning algorithm - KMeans to segment the species.

### Import relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

### Load the data.

```python
raw_data = pd.read_csv('iris_dataset.csv')
```

```python
raw_data.head()
```

```python
%matplotlib inline
plt.scatter(raw_data.sepal_length,raw_data.sepal_width)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

### Calculate the WCSS(Within Cluster Sum of Squares) 
### Elbow curve is used to determine the best no. of clusters required in a dataset.

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(i,n_init='auto', random_state=42)
    kmeans.fit_predict(raw_data)
    wcss.append(kmeans.inertia_)
wcss
```

### The cluster is mixing with each other so lets standardize the data = (mean-x)/std.dev

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x_scaled = scale.fit_transform(raw_data)
#x_scaled
```

```python
raw_data['Clusters'] = KMeans(2,n_init='auto', random_state=42).fit_predict(x_scaled)
raw_data
```

```python
%matplotlib inline
plt.scatter(raw_data.sepal_length,raw_data.sepal_width,c = raw_data['Clusters'], cmap = 'rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
raw_data['Clusters'] = KMeans(3,n_init='auto', random_state=42).fit_predict(x_scaled)
raw_data
```
